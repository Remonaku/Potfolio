```python
from glob import glob

train_img_list = glob('/content/drive/MyDrive/02_security_person/data_files/images/train/*.jpg')
test_img_list = glob('/content/drive/MyDrive/02_security_person/data_files/images/test/*.jpg')
val_img_list = glob('/content/drive/MyDrive/02_security_person/data_files/images/val/*.jpg')

print(len(train_img_list))
print(len(test_img_list))
print(len(val_img_list))
```

```python
with open('/content/drive/MyDrive/02_security_person/data_files/labels/test.txt', 'w') as f:
    f.write('\n'.join(test_img_list)+'\n')

with open('/content/drive/MyDrive/02_security_person/data_files/labels/train.txt', 'w') as f:
    f.write('\n'.join(train_img_list)+'\n')

with open('/content/drive/MyDrive/02_security_person/data_files/labels/val.txt', 'w') as f:
    f.write('\n'.join(val_img_list)+'\n')
```
